# Change Model to Average Model Price to Overcome Max. Factor Limit

```python
# Import the dataset

df = pd.read_csv('Scraped Cars.csv', index_col = 0)
df
```

```python
# Check the number of different models
len(set(df['Model']))
```

```python
# Gathering the means of numerical data for each model
means = df.groupby('Model').mean()
means
```

```python
# Creating new column 'Model'
means['Model'] = means.index
means
```

```python
#find the mean price of all models
list_mean_price = list(means['Price (£)'])
list_mean_price
```

```python
# Create a dictionary which pairs the models with their mean prices

# Rename the price for the following step
means = means.rename(columns ={'Price (£)': 'Price'})

# Create the dictionary pairing the models and mean prices
mean_price_dict = pd.Series(means.Price.values,index=means.Model).to_dict()
mean_price_dict
```

```python
#iterate through the dataframe, replacing the model with the mean model price
for i in range(len(df)):
    df['Model'].iloc[i] = mean_price_dict[df['Model'].iloc[i]]
```

```python
#check it has worked and rename the columns appropriately
df = df.rename(columns ={'Model': 'Mean Model Price (£)', 'Price (£)':'Actual Car Price (£)'})
df
```

```python
# Save to CSV
df.to_csv('Engineered Dataset.csv')
```

# Removing 9 Least Common Makes for Simple and Pruned Trees

```python
# re-import the dataset because returning to notebook and don't want to have to re-run the above code
df_top_makes = pd.read_csv('Engineered Dataset.csv', index_col = 0)
df_top_makes # print to ensure is correct
```

```python
values = df_top_makes['Make'].value_counts().keys().tolist()[:32] # Get the 32 most commonly occuring makes
print("32 most common makes:", values, "\n") #

df_top_makes = df_top_makes[df_top_makes['Make'].isin(values)] # Keep the makes that are in the above list only
df_top_makes
```

```python
# Save to CSV
df_top_makes.to_csv('Engineered Dataset for Simple and Pruned.csv')
```
